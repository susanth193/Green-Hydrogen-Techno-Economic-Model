```python
import numpy as np

i_data = np.arange(10,510,10) #current of range from 10 to 500
t = 3600 # 1 hour time
z = 2
f = 96485
mm = 2.016 # mass of hydrogen

H2_produced = np.round((i_data*t*mm)/(z*f),2)
print(H2_produced)
```

```python
import pandas as pd

V = 2 # voltage
P = (V*i_data)/1000 # power actual
n = 0.75 # efficiency
actual_H2 = H2_produced * n

data = {'Current':i_data, 'Hydrogen Produced':H2_produced, 'Power Input':P ,'SEC':np.round((P/actual_H2)*1000,2), 'Cost_per_Hour': P*7}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.plot(i_data, actual_H2, color='red', linestyle= "-", marker = "o", linewidth = 0.5, markersize = 2)
plt.title("The Production Curve")
plt.xlabel("Current")
plt.ylabel("Actual H2 Produced")
plt.grid(True)
plt.show()
```
